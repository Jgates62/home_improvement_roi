```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# Reading in data for both train.csv and test.csv

df = pd.read_csv('datasets/train.csv')
kag_test_df = pd.read_csv('datasets/test.csv')
```

```python
df.head()
```

## Data Cleaning & Feature Selection

```python
#Snake case all columns
df.columns =[col.replace(' ', '_').lower() for col in df.columns]
kag_test_df.columns =[col.replace(' ', '_').lower() for col in kag_test_df.columns]
```

```python
# Convert 'central_air' to binary
df['central_air'] = df['central_air'].map({'Y': 1, 'N': 0})
kag_test_df['central_air'] = kag_test_df['central_air'].map({'Y': 1, 'N': 0})
```

#### Create a column called 'qual_gr_liv_area' that will be the product of overall_qual and gr_liv_area

```python
df['qual_gr_liv_area'] = df['overall_qual'] * df['gr_liv_area']
kag_test_df['qual_gr_liv_area'] = kag_test_df['overall_qual'] * kag_test_df['gr_liv_area']
```

```python
# Take all the numerical columns from our train set, and impute the mean for any NAN values

# Get all columns that are numerical
features = df.select_dtypes(include= [np.number]).columns.tolist()
features.remove('saleprice') # And remove saleprice from our features

# For the 'fireplaces' column, replace null values with 0
df['fireplaces'].fillna(value= 0, inplace=True)
kag_test_df['fireplaces'].fillna(value= 0, inplace=True)

# Replace null values with the median of that column
[df[feat].fillna(value= df[feat].mean(), inplace=True) for feat in features]
[kag_test_df[feat].fillna(value= kag_test_df[feat].mean(), inplace=True) for feat in features];
```

```python
# set up a data frame with ONLY the features that you want and saleprice.

# create dummy columns for our categorical features
model_df = pd.get_dummies(df[cat_features], drop_first = True)
#model_df['saleprice'] = df['saleprice'] # Add the saleprice column
model_df.columns = [col.replace(' ', '_').lower() for col in model_df.columns ] #snake case the new dummy columns
# add the numeric columns to model_df
model_df = pd.merge(df[features], model_df, left_index= True, right_index = True)

# Now we do the same for kag_model_df
kag_model_df = pd.get_dummies(kag_test_df[cat_features], drop_first = True)
kag_model_df.columns = [col.replace(' ', '_').lower() for col in kag_model_df.columns ]
kag_model_df = pd.merge(kag_test_df[features], kag_model_df, left_index= True, right_index = True)
```

## Modeling

```python
#modelling_data = data.drop ('Shared Car', axis=1)
```

```python
#modelling_data = data.drop ('Shared Car', axis=1)
X = model_df
#X = model_df.drop('saleprice', axis = 1)
y = df['saleprice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
model_df.columns
```

```python
kag_model_df.columns
```

## Linear Regression

```python
# instantiate and fit

lr = LinearRegression()
lr.fit(X_train, y_train)
```

## KNN

```python
ss = StandardScaler()
ss.fit(X_train)
X_train_sc = ss.transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
knn = KNeighborsRegressor(n_neighbors = 5, weights = 'distance', p = 1)
```

```python
cross_val_score(knn, X_train_sc, y_train).mean() #For KNN, your predictions/test values (Accuracy)
```

```python
cross_val_score(knn, X_test_sc, y_test).mean()
```

```python
knn.fit(X_train_sc, y_train)
```

```python
knn.score(X_train_sc, y_train)
```

```python
knn.score(X_test_sc, y_test)
```

### Predictions & metrics

```python
# The predictions need to be done on the same features that we modeled on.
# Creating a variable to hold all our predictions

preds =lr.predict(kag_model_df)
train_preds =lr.predict(X)
train_residuals = y - train_preds
```

```python
#R2 Score
print('R2 scores: (train/test)', lr.score(X_train, y_train), lr.score(X_test, y_test))

#Cross val score
print('Cross val scores: (train/test)', cross_val_score(lr, X_train, y_train).mean(), cross_val_score(lr, X_test, y_test).mean() )

# MSE from training data
mse = metrics.mean_squared_error(y, train_preds)
print(f'Mean squared error: (train) {mse}')

# RMSE from training data
print(f'Root mean squared error: (train) {np.sqrt(mse)}')
```

### Create a function that will look at a house, and give one predicted sale price as is, and one predicted sale price with an additional home improvement.

```python
#sns.boxplot(x= 'heating', y = 'saleprice',data = df, palette='rainbow')
sns.boxplot(df['half_bath'], df['saleprice'], palette = 'viridis')
```

```python
plt.figure(figsize = (8, 4))
sns.boxplot(df['full_bath'], df['saleprice'], palette = 'viridis')
#plt.yticks(rotation = 45); # This rotates the x titles by 45 degrees
plt.title('Home Value by Number of Full Bathrooms', fontsize = 14)
plt.savefig('full_bath_boxplot.jpg', bbox_inches='tight')
```

```python
sns.boxplot(df['bsmt_full_bath'], df['saleprice'], palette = 'viridis')
```

```python
# For a particular feature, in this case 'full_bath', how would it affect the saleprice predictions
# if we were to add 1 full bath to every observation in our dataframe?

improvement = 'full_bath'

improved_df = model_df.copy()
orig_model_df = model_df.copy() #Save the model_df in it's current state so we can use it again

model_df[improvement] = df[improvement] # This will be our baseline dataframe
improved_df[improvement] = model_df[improvement] + 1  # This will be our "1 full bath added" dataframe


# get your predictions for model_df
model_preds = lr.predict(model_df)
model_df['pred_price'] = model_preds
model_df['saleprice'] = df['saleprice']

# get your predictions for improved_df
improved_preds = lr.predict(improved_df)
improved_df['pred_price'] = improved_preds
improved_df['saleprice'] = df['saleprice']
```

```python
model_df[[improvement, 'saleprice', 'pred_price']].head()
```

```python
improved_df[[improvement, 'saleprice', 'pred_price']].head()
```

```python
# What % improvement is this? (using the mean of our data)

(improved_df['pred_price'].mean() / model_df['pred_price'].mean()) - 1
```

#### What about if we keep full bathrooms the same, and add a half bathroom?

```python
improvement = 'half_bath'

model_df = orig_model_df.copy()
improved_df = orig_model_df.copy()

model_df[improvement] = df[improvement] # This will be our baseline dataframe
improved_df[improvement] = model_df[improvement] + 1  # This will be our "1 full bath added" dataframe


# get your predictions for model_df
model_preds = lr.predict(model_df)
model_df['pred_price'] = model_preds
model_df['saleprice'] = df['saleprice']

# get your predictions for improved_df
improved_preds = lr.predict(improved_df)
improved_df['pred_price'] = improved_preds
improved_df['saleprice'] = df['saleprice']
```

```python
# Here we actually see a very small decrease in the predictions whenwe added a half bath
(improved_df['pred_price'].mean() / model_df['pred_price'].mean()) - 1
```

#### What about if we added a basement full bath?

```python
improvement = 'bsmt_full_bath'

model_df = orig_model_df.copy()
improved_df = orig_model_df.copy()

model_df[improvement] = df[improvement] # This will be our baseline dataframe
improved_df[improvement] = model_df[improvement] + 1  # This will be our "1 full bath added" dataframe


# get your predictions for model_df
model_preds = lr.predict(model_df)
model_df['pred_price'] = model_preds
model_df['saleprice'] = df['saleprice']

# get your predictions for improved_df
improved_preds = lr.predict(improved_df)
improved_df['pred_price'] = improved_preds
improved_df['saleprice'] = df['saleprice']
```

```python
# Here we actually do see a 6% increase in mean saleprice when we add a basement full bath
(improved_df['pred_price'].mean() / model_df['pred_price'].mean()) - 1
```
