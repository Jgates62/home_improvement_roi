```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
```

```python
df = pd.read_csv("datasets/train.csv")
df.head()
```

### Data Cleaning

```python
#Snake case/rename column headers
df.columns = [col.replace(' ', '_').lower() for col in df.columns ]
```

```python
df['alley'].value_counts()
```

#### Here we need to figure out what values are missing and what we want to do with them.

```python
pd.set_option('display.max_rows', 500)
#print(df.isnull().sum())

#lot_frontage: replace null 0
df['lot_frontage'].mask(df['lot_frontage'].isnull(), 0, inplace = True)


#alley: replace null with 'naa'
#        NA 	No alley access -> naa: No alley access
df['alley'].mask(df['alley'].isnull(), 'naa', inplace = True)


#mas_vnr_type: Masonry veneer type
#         None	None           -> no_ven : No masonry veneer
df['mas_vnr_type'].mask(df['mas_vnr_type'].isnull(), 'no_ven', inplace = True)

#mas_vnr_area: Masonry veneer area in square feet
#        replace null with 0
df['mas_vnr_area'].mask(df['mas_vnr_area'].isnull(), 0, inplace = True)

# bsmt_qual: Evaluates the height of the basement
#        NA	No Basement                     -> no_bsmt : No basement
df['bsmt_qual'].mask(df['bsmt_qual'].isnull(), 0, inplace = True)

# bsmt_cond: Evaluates the general condition of the basement
#        NA	No Basement               -> no_bsmt : No basement
df['bsmt_cond'].mask(df['bsmt_cond'].isnull(), 'no_bsmt', inplace = True)

# bsmt_exposure: Refers to walkout or garden level walls
#        NA	No Basement                 -> no_bsmt : No basement
df['bsmt_exposure'].mask(df['bsmt_exposure'].isnull(), 'no_bsmt', inplace = True)

# bsmtfin_type_1
# impute NA	No Basement       -> no_bsmt : No basement
df['bsmtfin_type_1'].mask(df['bsmtfin_type_1'].isnull(), 'no_bsmt', inplace = True)

# bsmtfin_sf_1 : Type 1 finished square feet
# replace null with 0
df['bsmtfin_sf_1'].mask(df['bsmtfin_sf_1'].isnull(), 0, inplace = True)

# bsmtfin_type_2
# impute NA	No Basement       -> no_bsmt : No basement
df['bsmtfin_type_2'].mask(df['bsmtfin_type_2'].isnull(), 'no_bsmt', inplace = True)

# bsmtfin_sf_2
# replace nan with 0
df['bsmtfin_sf_2'].mask(df['bsmtfin_sf_2'].isnull(), 0, inplace = True)

# bsmt_unf_sf           
# replace nan with 0
df['bsmt_unf_sf'].mask(df['bsmt_unf_sf'].isnull(), 0, inplace = True)

# total_bsmt_sf         
# replace nan with 0
df['total_bsmt_sf'].mask(df['total_bsmt_sf'].isnull(), 0, inplace = True)

# bsmt_full_bath : Basement full bathrooms
# replace nan with 0
df['bsmt_full_bath'].mask(df['bsmt_full_bath'].isnull(), 0, inplace = True)

# bsmt_half_bath : Basement half bathrooms
# replace nan with 0
df['bsmt_half_bath'].mask(df['bsmt_half_bath'].isnull(), 0, inplace = True)

# fireplace_qu : Fireplace quality
#     NA	No Fireplace       -> impute no_fire for NA
df['fireplace_qu'].mask(df['fireplace_qu'].isnull(), 'no_fire', inplace = True)

# garage_type: Garage location
#         NA	No Garage         -> impute no_gar for NA
df['garage_type'].mask(df['garage_type'].isnull(), 'no_gar', inplace = True)

# garage_yr_blt: Year garage was built
# replace nan with 0
df['garage_yr_blt'].mask(df['garage_yr_blt'].isnull(), 0, inplace = True)

# garage_finish
#     NA	No Garage     -> impute no_gar for NA
df['garage_finish'].mask(df['garage_finish'].isnull(), 0, inplace = True)

# garage_cars : Size of garage in car capacity
#     -> replace nan with 0
df['garage_cars'].mask(df['garage_cars'].isnull(), 0, inplace = True)

# garage_area : Size of garage in square feet
#     -> replace nan with 0
df['garage_area'].mask(df['garage_area'].isnull(), 0, inplace = True)

# garage_qual
#         NA	No Garage         -> impute no_gar for NA
df['garage_qual'].mask(df['garage_qual'].isnull(), 'no_gar', inplace = True)

# garage_cond
#         NA	No Garage         -> impute no_gar for NA
df['garage_cond'].mask(df['garage_cond'].isnull(), 'no_gar', inplace = True)

# pool_qc
#     NA	No Pool    -> impute no_pool for NA
df['pool_qc'].mask(df['pool_qc'].isnull(), 'no_pool', inplace = True)

# fence
#     NA	No Fence    -> impute no_fen for NA
df['fence'].mask(df['fence'].isnull(), 'no_fen', inplace = True)

# misc_feature
#     NA	None       -> replace nan with no_misc
df['misc_feature'].mask(df['misc_feature'].isnull(), 'no_misc', inplace = True)

print(df.isnull().sum())
df.head()
```
