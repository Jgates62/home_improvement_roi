```python
#Snake case/rename column headers
df.columns = [col.replace(' ', '_').lower() for col in df.columns ]
```

```python
bld_typ_df = pd.get_dummies(df['bldg_type'], prefix = None)
bld_typ_df.columns = [col.replace(' ', '_').lower() for col in bld_typ_df.columns ]
bld_typ_df.columns
```

```python
h_style_df = pd.get_dummies(df['house_style'], prefix = None)
h_style_df.columns = [col.replace(' ', '_').lower() for col in h_style_df.columns ]
h_style_df.columns
```

```python
df = pd.merge(df, zone_df, left_index= True, right_index = True)
df = pd.merge(df, utility_df, left_index= True, right_index = True)
df = pd.merge(df, neighborhood_df, left_index= True, right_index = True)
df = pd.merge(df, bld_typ_df, left_index= True, right_index = True)
df = pd.merge(df, h_style_df, left_index= True, right_index = True)
df.head(2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42)
#X_train, X_test, y_train, y_test = train_test_split(X, y) # Without random state
```

```python
#instantiate and fit
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#R2 Score
print('r2 scores: ', lr.score(X_train, y_train), lr.score(X_test, y_test))

#Cross val score
print('cross val scores: ', cross_val_score(lr, X_train, y_train).mean(), cross_val_score(lr, X_test, y_test).mean() )
```

```python
preds = lr.predict(X_test)

residuals = y_test - preds
```
